**SPLIT THE DATA**

**LINEAR REGRESSION MODEL**

```python
from sklearn.linear_model import LinearRegression
```

```python
x_train, y_train = train_data.drop(columns = ["median_house_value"]), train_data["median_house_value"]
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
test_data = x_test.join(y_test)

test_data["total_rooms"] = np.log(test_data["total_rooms"] + 1)
test_data["total_bedrooms"] = np.log(test_data["total_bedrooms"] + 1)
test_data["population"] = np.log(test_data["population"] + 1)
test_data["households"] = np.log(test_data["households"] + 1)

test_data = test_data.join(pd.get_dummies(test_data.ocean_proximity)).drop(columns = ["ocean_proximity"])

test_data["bedroom_ratio"] = test_data["total_bedrooms"] / test_data["total_rooms"]
test_data["household_rooms"] = test_data["total_rooms"] / test_data["households"]
```

```python
x_test, y_test = test_data.drop(columns = ["median_house_value"]), test_data["median_house_value"]
```

```python
x_train
```

```python
model.score(x_test, y_test)
```

**Random Forest Model**

```python
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor()

forest.fit(x_train, y_train)
```

```python
forest.score(x_test, y_test)
```

```python
from sklearn.model_selection import GridSearchCV

forest = RandomForestRegressor()

param_grid = {
    "n_estimators": [30, 50, 100],
    "max_features": [8, 12, 20],
    "min_samples_split": [2, 4, 6, 8]
}

grid_search = GridSearchCV(forest, param_grid, cv=5, scoring="neg_mean_squared_error", return_train_score=True)

grid_search.fit(x_train, y_train)
```

```python
best_forest = grid_search.best_estimator_
best_forest
```

```python
best_forest.score(x_test, y_test)
```

```python
predict = best_forest.predict(x_test)
```
